#### Importing necessary libraries

```python
import numpy as np
import pandas as pd 
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('admission_data.csv')
```

```python
df.info()
```

#### Data outlines

* Histplot

```python
num_list=['GRE Score', 'TOEFL Score', 'University Rating', 'SOP', 'LOR ', 'CGPA',
       'Research', 'Chance of Admit ']
fig = plt.figure(figsize=(10,10))

for i in range(len(num_list)):
    plt.subplot(3,3,i+1)
    plt.title(num_list[i])
    sns.histplot(df[num_list[i]])

plt.tight_layout()
```

* Boxplot

```python
fig = plt.figure(figsize=(10,10))

for i in range(len(num_list)):
    plt.subplot(3,3,i+1)
    plt.title(num_list[i])
    plt.boxplot(df[num_list[i]])

plt.tight_layout()
```

* Pairplot

```python
sns.pairplot(df)
```

#### Correlation analysis

```python
plt.figure(figsize = (15,8))
sns.heatmap(df.corr(),annot=True, cbar=True, cmap='Blues', fmt='.1f')
```

Chance of admit has strong positive correlation with
* CGPA 0.9)
* GRE Score(0.8)
* TOEFL Score(0.8)
* University Rating(0.7)
* SOP(0.7)
